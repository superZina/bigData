```python
#성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다.
#나이를 Age로 하면 그룹이 많아지기 때문에 Age_Class로 대체
df1 = df.groupby(['Sex','Age_class','Pclass'])['Survived'].mean()
```

```python
df1
```

```python
df2 = df1.unstack('Pclass')
#Pclass를 위로 올려줌
df2
#결과를 보면 1등객실의 미성년,성년 여자가 가장 생존률이 높고, 두번째로 1등객식의 미성년,성년 남자가 생존률이 높다. 
```

```python
#피봇테이블 
#성별, 선실에 의한 생존률을 피봇데이터 형태로 볼 수 있다.
df1= df.pivot_table(['Survived'],index = ['Sex','Pclass','Age_class'],aggfunc = 'mean')
df2 = df.pivot_table(['Survived'],index = ['Sex','Pclass'],aggfunc = 'mean')
df1
#groupby와 다른점은 NaN값은 아예 테이블에 포함시키지 않는다는것!
```
